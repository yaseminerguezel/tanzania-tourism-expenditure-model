```python
'''dummy variables
'''
age_with_dummy = pd.get_dummies(newdf['age_group'],drop_first=True)
data_all = pd.concat([newdf,age_with_dummy],axis=1)


```

```python
travel_with_dummy = pd.get_dummies(newdf['travel_with'],drop_first=True)
data_all1 = pd.concat([data_all,travel_with_dummy],axis=1)
```

```python
info_with_dummy = pd.get_dummies(newdf['info_source'],drop_first=True)
data_all4 = pd.concat([data_all3,info_with_dummy],axis=1)
```

```python
'''we replaced the yes and no with 1 and 0 for all the packages
'''
data_all7 = data_all6.replace({'Yes':1, 'No': 0})
```

```python
data_all8 = data_all7.replace({'SWIZERLAND':1, 'UNITED KINGDOM': 2 ,'CHINA':3,'DENMARK':4,'UNITED STATES OF AMERICA':5,'ITALY':6,'FRANCE':7,'AUSTRALIA':8,'SOUTH AFRICA':9,'GERMANY':10,'SPAIN':11,'CANADA':12,'NETHERLANDS':13})
```

```python
###removing outliers Inter-quartile Range (IQR)
###IQR = Q3-Q1
###Lower Boundary = Q1-1.5*IQR
###Upper Boundary = Q3+1.5*IQR

```

```python
###removing outliers Inter-quartile Range (IQR)
###IQR = Q3-Q1
###Lower Boundary = Q1-1.5*IQR
###Upper Boundary = Q3+1.5*IQR

def outliers(df,ft):
    Q1 = df[ft].quantile(0.25)
    Q3 = df[ft].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR

    ls = df.index[(df[ft] < lower_bound)| (df[ft] > upper_bound)]
    return ls
```

```python
#create an empty list to store the output indices from multiple columns
index_list = []
for feature in ['night_mainland','night_zanzibar','total_cost']:
    index_list.extend(outliers(data_all8,feature))
```

```python
def remove(df,ls):
    ls = sorted(set(ls))
    df = df.drop(ls)
    return df
```

```python
df_cleaned =remove(data_all8,index_list)
```
